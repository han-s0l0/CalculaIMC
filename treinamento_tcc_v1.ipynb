```python
#converte png pra jpg
import cv2
import shutil
import os

imagens = '/content/drive/MyDrive/DatasetPrincipal/train/images/'

lisImgs = os.listdir(imagens)

uno = 0

for i in lisImgs:
  nome = i.split(".")[0]
  arq = i.split(".")[1]
  if arq == "PNG" or arq == "png":
    imgpng = cv2.imread(imagens + i)
    try:
      cv2.imwrite(imagens + nome + ".jpg", imgpng, [int(cv2.IMWRITE_JPEG_QUALITY), 100])
      os.remove(imagens + i)
      uno += 1
      print(uno)

    except:
      print("deu erro")
```
